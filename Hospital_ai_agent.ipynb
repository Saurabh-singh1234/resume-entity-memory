```python
# ===========================
# ADVANCED HOSPITAL AI AGENT
# SINGLE CELL COLAB PROJECT
# ===========================

!pip -q install pandas matplotlib

import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import random

# ---------------------------
# DATABASES
# ---------------------------

patients = []
appointments = []
bills = []

doctors = {
    "Cardiology":"Dr. Sharma",
    "Neurology":"Dr. Verma",
    "Orthopedic":"Dr. Singh",
    "General":"Dr. Khan"
}

medicine_stock = {
    "Paracetamol":120,
    "Crocin":80,
    "Amoxicillin":50,
    "Vitamin C":100
}

# ---------------------------
# FUNCTIONS
# ---------------------------

def register_patient():
    name=input("Patient Name: ")
    age=int(input("Age: "))
    gender=input("Gender(M/F): ")

    pid=f"P{random.randint(1000,9999)}"

    patients.append({
        "ID":pid,
        "Name":name,
        "Age":age,
        "Gender":gender
    })

    print(f"\nPatient Registered Successfully")
    print("Patient ID:",pid)

def book_appointment():

    if len(patients)==0:
        print("No patient registered")
        return

    pid=input("Patient ID: ")

    print("\nDepartments:")
    for d in doctors:
        print("-",d)

    dept=input("\nDepartment: ")

    doctor=doctors.get(dept,"Dr. Khan")

    appointments.append({
        "Patient_ID":pid,
        "Department":dept,
        "Doctor":doctor
    })

    print("\nAppointment Booked")
    print("Doctor Assigned:",doctor)

def generate_bill():

    pid=input("Patient ID: ")

    consultation=500
    medicine=300
    test=700

    total=consultation+medicine+test

    bills.append({
        "Patient_ID":pid,
        "Amount":total
    })

    print("\nBill Generated")
    print("Total Amount =",total)

def medicine_inventory():

    print("\nMedicine Inventory\n")

    for med,qty in medicine_stock.items():
        print(med,"=",qty)

def ai_report_summary():

    report=input("\nPaste Medical Report:\n")

    words=len(report.split())

    print("\n===== AI SUMMARY =====")
    print("Report Length:",words,"words")
    print("Risk Level:",
          "HIGH" if "critical" in report.lower()
          else "MEDIUM" if "abnormal" in report.lower()
          else "LOW")

def dashboard():

    if len(patients)==0:
        print("No Data")
        return

    gender_df=pd.DataFrame(patients)

    gender_count=gender_df["Gender"].value_counts()

    plt.figure(figsize=(5,4))
    gender_count.plot(kind="bar")
    plt.title("Patient Gender Distribution")
    plt.show()

    if len(bills)>0:

        bill_df=pd.DataFrame(bills)

        plt.figure(figsize=(5,4))
        bill_df["Amount"].plot(kind="hist")
        plt.title("Revenue Distribution")
        plt.show()

# ---------------------------
# MAIN AGENT
# ---------------------------

print("="*50)
print("🏥 ADVANCED HOSPITAL AI AGENT")
print("="*50)

while True:

    print("""
1 Register Patient
2 Book Appointment
3 Generate Bill
4 Medicine Inventory
5 AI Report Analysis
6 Dashboard
7 View Patients
8 Exit
""")

    choice=input("Choose Option: ")

    if choice=="1":
        register_patient()

    elif choice=="2":
        book_appointment()

    elif choice=="3":
        generate_bill()

    elif choice=="4":
        medicine_inventory()

    elif choice=="5":
        ai_report_summary()

    elif choice=="6":
        dashboard()

    elif choice=="7":
        print(pd.DataFrame(patients))

    elif choice=="8":
        print("Hospital AI Agent Closed")
        break

    else:
        print("Invalid Choice")
```
