```python
# =========================
# TRAVEL + VOICE AGENT (ONE CELL)
# =========================

!pip -q install SpeechRecognition gTTS pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import speech_recognition as sr
from gtts import gTTS
from IPython.display import Audio, display
from google.colab import files
import os

# ================= VOICE OUTPUT =================
def speak(text):
    print("🤖:", text)
    tts = gTTS(text)
    file = "voice.mp3"
    tts.save(file)
    return Audio(file, autoplay=True)

# ================= VOICE INPUT =================
def listen():
    r = sr.Recognizer()

    try:
        with sr.Microphone() as source:
            print("🎤 Speak now...")
            audio = r.listen(source)
            text = r.recognize_google(audio)
            print("🧑 You said:", text)
            return text.lower()
    except:
        print("⚠️ Voice input not working in Colab, using text input instead")
        return input("Type your command: ").lower()

# ================= UPLOAD FILE =================
print("Upload your travel dataset CSV")
uploaded = files.upload()

file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

print("\nDataset Loaded")
print(df.head())

# ================= CLEAN COLUMN NAMES =================
df.columns = [c.strip().lower() for c in df.columns]

# flexible column detection
gender_col = [c for c in df.columns if "gender" in c][0]
country_col = [c for c in df.columns if "country" in c or "nationality" in c][0]
hotel_col = [c for c in df.columns if "hotel" in c or "accommodation" in c][0]
budget_col = [c for c in df.columns if "cost" in c or "budget" in c][0]

# ================= ANALYSIS =================
gender_count = df[gender_col].value_counts()
country_count = df[country_col].value_counts().head(10)
hotel_count = df[hotel_col].value_counts().head(10)

df[budget_col] = pd.to_numeric(df[budget_col], errors="coerce")

# ================= GRAPHS =================
plt.figure()
gender_count.plot(kind="bar", color=["blue","pink"])
plt.title("Male vs Female Travelers")
plt.show()

plt.figure()
country_count.plot(kind="bar")
plt.title("Top Countries")
plt.show()

plt.figure()
hotel_count.plot(kind="bar", color="green")
plt.title("Top Hotels")
plt.show()

plt.figure()
sns.histplot(df[budget_col].dropna(), bins=10, kde=True)
plt.title("Budget Distribution")
plt.show()

# ================= SAVE FILES =================
gender_count.to_csv("gender_analysis.csv")
country_count.to_csv("country_analysis.csv")
hotel_count.to_csv("hotel_analysis.csv")
df[[budget_col]].to_csv("budget_data.csv")

# ================= VOICE TRAVEL AGENT =================
def travel_agent():
    display(speak("Travel voice agent is ready"))

    while True:
        display(speak("Say country, hotel, gender, budget or exit"))

        q = listen()

        if "exit" in q:
            display(speak("Goodbye"))
            break

        # ================= SMART DETECTION =================

        # GENDER
        if q in ["male", "female"]:
            display(speak("Gender data"))
            # Convert to string to handle potential NaNs before string operations
            print(df[df[gender_col].astype(str).str.lower() == q])

        # COUNTRY (ANY TEXT LIKE INDIA / USA / INDIAN)
        elif any(c in q for c in ["india", "usa", "uk", "america"]):
            display(speak("Country data"))
            # Convert to string to handle potential NaNs before string operations
            print(df[df[country_col].astype(str).str.lower().str.contains(q)])

        # HOTEL (villa/taj/hilton etc.)
        # Check if the query 'q' is a substring of any non-NaN unique hotel names.
        elif any(q in hotel_name for hotel_name in df[hotel_col].astype(str).str.lower().unique() if hotel_name != 'nan'):
            display(speak("Hotel data"))
            # Convert to string to handle potential NaNs before string operations
            print(df[df[hotel_col].astype(str).str.lower().str.contains(q)])

        # BUDGET (NUMBER INPUT)
        elif q.replace(" ", "").isdigit():
            display(speak("Budget filter applied"))
            print(df[df[budget_col] <= int(q)])

        else:
            display(speak("No match found. Try country, hotel, gender or budget"))

# RUN AGENT
travel_agent()

# ================= DOWNLOAD FILES =================
print("\nDownloading result files...")
files.download("gender_analysis.csv")
files.download("country_analysis.csv")
files.download("hotel_analysis.csv")
files.download("budget_data.csv")
```
