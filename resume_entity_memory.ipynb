```python
class EntityMemory:

    def __init__(self):
        self.entities = {}

    def add_candidate(self,name,skills,experience):

        self.entities[name] = {
            "skills": skills,
            "experience": experience
        }

    def show_memory(self):
        return self.entities
```

```python
import os

memory = EntityMemory()

resume_folder = "resumes"

for file in os.listdir(resume_folder):

    path = os.path.join(resume_folder,file)

    with open(path,"r") as f:

        data = f.read()

    lines = data.split("\n")

    name = ""
    skills = []
    experience = ""

    collect_skills = False

    for line in lines:

        line = line.strip()

        if line.startswith("Name:"):
            name = line.replace("Name:","").strip()

        elif line.startswith("Skills"):
            collect_skills = True

        elif line.startswith("Experience"):
            collect_skills = False

        elif "Year" in line:
            experience = line

        elif collect_skills and line:
            skills.append(line.lower())

    memory.add_candidate(
        name,
        skills,
        experience
    )
```

```python
memory.show_memory()
```

```python
with open("jobdescription.txt","r") as f:
    jd_text = f.read().lower()

required_skills = [
    "python",
    "langchain",
    "llm",
    "rag"
]
```

```python
class ResumeScreeningAgent:

    def __init__(self,memory):
        self.memory = memory

    def rank_candidates(self):

        results = []

        for candidate,data in self.memory.entities.items():

            matched = []

            for skill in required_skills:

                if skill in data["skills"]:
                    matched.append(skill)

            score = len(matched)

            results.append({
                "candidate":candidate,
                "score":score,
                "matched":matched
            })

        results.sort(
            key=lambda x:x["score"],
            reverse=True
        )

        return results
```

```python
agent = ResumeScreeningAgent(memory)

ranking = agent.rank_candidates()

for r in ranking:

    print("Candidate:",r["candidate"])
    print("Score:",r["score"])
    print("Matched Skills:",r["matched"])
    print("-"*40)
```

```python
best = ranking[0]

print("Selected Candidate")
print(best["candidate"])
```
